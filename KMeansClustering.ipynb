```python
from sklearn.datasets import load_iris
a=load_iris()
x=a.data
y=a.target
from sklearn.cluster import KMeans
c=KMeans(init="k-means++",n_clusters=3,n_init=10,random_state=0).fit(x)
p=c.fit_predict(x)
print("Actual values")
print(y)
print("Predicted values")
print(p)
```

```python
import matplotlib.pyplot as plt
import numpy
plt.scatter(x[p==0,0],x[p==0,1],marker="*",c="red",label="Iris-setosa")
plt.scatter(x[p==1,0],x[p==1,1],marker="*",c="blue",label="Iris-versicolor")
plt.scatter(x[p==2,0],x[p==2,1],marker="*",c="black",label="Iris-verginica")
plt.scatter(c.cluster_centers_[:,0],c.cluster_centers_[:,1],s=100,c="green",label="Centroids")
plt.legend()
```

```python
cen=c.cluster_centers_
print("Centroids")
print(cen)
```
