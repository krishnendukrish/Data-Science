```python
from sklearn.datasets import load_iris
import numpy as np
x,y=load_iris(return_X_y=True)
x=x[:,np.newaxis,2]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=.25)
```

```python
from sklearn.linear_model import LinearRegression
c=LinearRegression()
c.fit(xtrain,ytrain)
p=c.predict(xtest)
print("Actual values")
print(ytest)
print("Predicted values")
print(p)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2=r2_score(ytest,p)
print("r2_score")
print(r2)
```

```python
mean=mean_squared_error(ytest,p)
print("mean_squared_error")
print(mean)
```

```python
import matplotlib.pyplot as plt
print("Graphical Representation of Linear Reression")
plt.scatter(xtest,ytest,color="green")
plt.plot(xtest,p)
plt.title("Actal v/s Predicted")
plt.show()
```

```python
print("Coefficient")
print(c.coef_)
```

```python
print("Intercept")
print(c.intercept_)
```
